```python
from sklearn.model_selection import train_test_split
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Single Linear Regression
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
pred=LR.predict(X_test)
pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error
lin_pred_train=LR.predict(X_train)
```

```python
print("MAE",mean_absolute_error(lin_pred_train,y_train))
print("MSE",mean_squared_error(lin_pred_train,y_train))
print("RMSE",root_mean_squared_error(lin_pred_train,y_train))
```
